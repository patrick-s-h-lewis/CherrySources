```python
import os
import json
```

```python
dirs = os.listdir('data')
total=0
total_urls = []
for di in dirs:
    with open('data/'+di+'/urls.json','r')as f:
        print(di)
        j = json.load(f)
        j.remove({})
        total_urls+=j
        print(len(j))
        total+=len(j)
print(total)
len(total_urls)
with open('scrape_these.json','w')as f:
    json.dump(total_urls,f,indent=4, separators=(',', ': '))
```
